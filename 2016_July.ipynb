```python
import pandas as pd
```

```python
# Loadind the csv files
df_csv1 = pd.read_csv('201607-citibike-tripdata_1.csv')
df_csv2 = pd.read_csv('201607-citibike-tripdata_2.csv')
```

```python
# Concatenate the DataFrames
combined_df = pd.concat([df_csv1, df_csv2])
combined_df.tail()
```

```python
combined_df.shape
```

```python
# replace 0,1,2 in the gender column to 'NA', 'Male', 'Female'
combined_df['gender'] = combined_df['gender'].replace({0: 'Other', 1: 'Male', 2: 'Female'})
combined_df.head()
```

```python
# drop all NaN values in the combined_df dataframe
combined_df.dropna(inplace=True)
combined_df.info()
```

```python
# calculate the age by subtracting the birth year from the year of the data i.e. 2016
combined_df['Age'] = 2016 - combined_df['birth year']
combined_df.drop(['birth year'], axis=1, inplace=True)
combined_df.head()
```

```python
# dropping rows where Age is greater than 75
combined_df = combined_df[combined_df['Age'] <= 75]
```

```python
# Convert the starttime & stoptime column to a datetime format
combined_df['starttime'] = pd.to_datetime(combined_df['starttime'], format='%m/%d/%Y %H:%M:%S')
combined_df['stoptime'] = pd.to_datetime(combined_df['starttime'], format='%m/%d/%Y %H:%M:%S')
```

```python
combined_df.info()
```

```python
combined_df.to_csv('2016_July.csv', index=False)
```
